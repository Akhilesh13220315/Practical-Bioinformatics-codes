```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load and clean dataset
df = pd.read_csv("diabetes.csv")
df = df.dropna()

# ==========================
# Remove mean, mode, median rows (optional logic)
# ==========================
def remove_statistical_rows(df, group_col='Outcome'):
    to_remove = []
    for outcome in df[group_col].unique():
        group_df = df[df[group_col] == outcome]
        for col in df.columns:
            if col == group_col:
                continue
            col_mean = group_df[col].mean()
            col_median = group_df[col].median()
            col_mode = group_df[col].mode()[0] if not group_df[col].mode().empty else np.nan

            # Remove rows where value == mean OR median OR mode (rounded to avoid float issues)
            to_remove.extend(group_df[
                (group_df[col].round(2) == round(col_mean, 2)) |
                (group_df[col].round(2) == round(col_median, 2)) |
                (group_df[col].round(2) == round(col_mode, 2))
            ].index)

    df_cleaned = df.drop(set(to_remove))
    return df_cleaned

df = remove_statistical_rows(df)

print("‚úÖ After removing mean, mode, median rows:")
print(df.head())

# ==========================
# Machine Learning Model
# ==========================
X = df.drop("Outcome", axis=1)
y = df["Outcome"]

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions with probabilities
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]  # probability of class = 1 (Diabetes)

# Evaluation
print("üìä Classification Report:")
print(classification_report(y_test, y_pred))

# Visualize confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ==========================
# Visualize Detection Percentages (Probabilities)
# ==========================
result_df = pd.DataFrame(X_test, columns=X.columns)
result_df["Actual"] = y_test.values
result_df["Predicted"] = y_pred
result_df["Diabetes_Probability"] = y_prob * 100

# Animated bubble chart of predicted probabilities
fig = px.scatter(
    result_df,
    x="Age",
    y="BMI",
    size="Diabetes_Probability",
    color="Predicted",
    hover_data=["Diabetes_Probability"],
    animation_frame=result_df["Diabetes_Probability"].round(-1),  # Group by 10s
    title="üß† Predicted Diabetes Risk Bubble Chart (Age vs BMI)",
    labels={"Predicted": "Predicted Diabetes (0=No, 1=Yes)"}
)
fig.show()
```

```python
# Install Biopython if you don't have it (uncomment the line below in Colab)
# !pip install biopython

from Bio import Entrez
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction
from Bio import Align # For sequence alignment

# Set your email for NCBI Entrez API access
# It's good practice to use a real email.
Entrez.email = "your.email@example.com" # IMPORTANT: Replace with your actual email

def perform_basic_sequence_analysis_from_ncbi(accession_id):
    """
    Fetches a nucleotide sequence from NCBI's GenBank database using its accession ID,
    and performs basic sequence analysis (length, GC content, transcription, translation).

    Args:
        accession_id (str): The accession ID of the sequence (e.g., "NM_000546.6" for human TP53 mRNA).
    """
    print(f"\n--- Basic Analysis for Accession ID: {accession_id} ---")
    try:
        # Step 1: Fetch the sequence record from NCBI Nucleotide database
        # rettype="gb" for GenBank format, retmode="text" for plain text
        handle = Entrez.efetch(db="nuccore", id=accession_id, rettype="gb", retmode="text")
        record = SeqIO.read(handle, "genbank")
        handle.close()

        print(f"Sequence ID: {record.id}")
        print(f"Description: {record.description}")
        print(f"Organism: {record.annotations.get('organism', 'N/A')}")
        print(f"Molecule Type: {record.annotations.get('molecule_type', 'N/A')}")

        # Extract the sequence object
        dna_sequence = record.seq
        print(f"\n--- Basic Sequence Properties ---")

        # 2. Get sequence length
        seq_length = len(dna_sequence)
        print(f"Length of DNA sequence: {seq_length} bp")

        # 3. Calculate GC content
        gc_content_percentage = gc_fraction(dna_sequence) * 100
        print(f"GC content: {gc_content_percentage:.2f}%")

        # 4. Transcribe DNA to RNA
        # Biopython's .transcribe() method directly converts DNA to RNA
        rna_sequence = dna_sequence.transcribe()
        print(f"RNA sequence (first 50 bases): {rna_sequence[:50]}...")

        # 5. Translate RNA to Protein
        # Biopython's .translate() method translates RNA to protein.
        # It handles stop codons by default (represented as '*')
        protein_sequence = rna_sequence.translate()
        print(f"Protein sequence (first 50 amino acids): {protein_sequence[:50]}...")

        # Optional: Check for stop codons in the protein sequence
        if '*' in protein_sequence:
            print(f"Note: Stop codons detected in the protein sequence (indicated by '*')")
            # For a more precise protein, you'd typically extract the Coding Sequence (CDS)
            # feature from record.features and translate only that specific part.

    except Exception as e:
        print(f"An error occurred during basic sequence analysis for {accession_id}: {e}")
        print("Please check the accession ID and your internet connection.")

def perform_pairwise_sequence_alignment(accession_id1, accession_id2):
    """
    Fetches two nucleotide sequences from NCBI and performs pairwise global and local alignment.

    Args:
        accession_id1 (str): Accession ID of the first sequence.
        accession_id2 (str): Accession ID of the second sequence.
    """
    print(f"\n--- Performing Pairwise Sequence Alignment ---")
    print(f"Comparing {accession_id1} and {accession_id2}")

    try:
        # Fetch first sequence
        handle1 = Entrez.efetch(db="nuccore", id=accession_id1, rettype="fasta", retmode="text")
        seq1_record = SeqIO.read(handle1, "fasta")
        handle1.close()
        seq1 = seq1_record.seq

        # Fetch second sequence
        handle2 = Entrez.efetch(db="nuccore", id=accession_id2, rettype="fasta", retmode="text")
        seq2_record = SeqIO.read(handle2, "fasta")
        handle2.close()
        seq2 = seq2_record.seq

        # Initialize the aligner
        # PairwiseAligner uses a global alignment algorithm (Needleman-Wunsch by default)
        # and can also perform local alignment (Smith-Waterman).
        # Adjust gap_open and gap_extend penalties as needed for your specific analysis.
        aligner = Align.PairwiseAligner()
        aligner.match_score = 2
        aligner.mismatch_score = -1
        aligner.open_gap_score = -0.5
        aligner.extend_gap_score = -0.1

        # --- Global Alignment ---
        print("\n--- Global Alignment (Needleman-Wunsch) ---")
        # Get all possible global alignments
        global_alignments = aligner.align(seq1, seq2)

        # Print the best global alignment(s)
        if global_alignments:
            best_global_alignment = max(global_alignments, key=lambda x: x.score)
            print(f"Score: {best_global_alignment.score}")
            print(best_global_alignment) # This prints the alignment visually
        else:
            print("No global alignment found.")

        # --- Local Alignment ---
        print("\n--- Local Alignment (Smith-Waterman) ---")
        aligner.mode = 'local' # Set aligner to local mode
        local_alignments = aligner.align(seq1, seq2)

        # Print the best local alignment(s)
        if local_alignments:
            best_local_alignment = max(local_alignments, key=lambda x: x.score)
            print(f"Score: {best_local_alignment.score}")
            print(best_local_alignment) # This prints the alignment visually
        else:
            print("No local alignment found.")

    except Exception as e:
        print(f"An error occurred during pairwise alignment: {e}")
        print("Please check the accession IDs and your internet connection.")

def conceptual_genome_comparison_explanation():
    """
    Provides a conceptual explanation of genome comparison, as full genome alignment
    is computationally intensive for a simple script.
    """
    print("\n--- Conceptual Explanation: Genome Comparison ---")
    print("Directly performing full genome-to-genome alignment in a simple script like this")
    print("is computationally very intensive and requires specialized tools and significant")
    print("computational resources (e.g., HPC clusters or cloud computing).")
    print("\nHowever, genome comparison is a crucial part of bioinformatics, used for:")
    print("  - Identifying evolutionary relationships between species.")
    print("  - Finding conserved regions (e.g., genes, regulatory elements) across genomes.")
    print("  - Detecting large-scale structural variations (insertions, deletions, rearrangements).")
    print("  - Understanding pathogen evolution and outbreaks.")
    print("\nCommon approaches and specialized tools for genome comparison include:")
    print("1.  **Whole-Genome Alignment Tools:**")
    print("    -   **MUMmer:** Designed for aligning large prokaryotic and eukaryotic genomes.")
    print("    -   **LASTZ / MAFFT:** Used for aligning large genomic regions or entire chromosomes.")
    print("    -   These tools often generate output that highlights synteny (conservation of gene order).")
    print("2.  **Sequence Similarity Search (e.g., BLAST at Genome Scale):**")
    print("    -   Using tools like BLAST to search a query sequence (e.g., a gene) against an entire genome database.")
    print("    -   While not a full alignment, it identifies homologous regions and is a form of comparison.")
    print("3.  **Genome Browsers (e.g., UCSC Genome Browser, Ensembl):**")
    print("    -   Interactive web-based tools that allow visual comparison of multiple genomes,")
    print("        displaying aligned regions, gene annotations, and variant tracks.")
    print("    -   They often integrate data from various alignment and annotation tools.")
    print("4.  **Comparative Genomics Pipelines:**")
    print("    -   More complex pipelines that integrate multiple tools for tasks like gene ortholog identification,")
    print("        phylogenetic tree construction based on whole genomes, and detection of gene family expansions/contractions.")
    print("\nFor practical purposes in a research or diagnostic setting, you would typically use these specialized")
    print("software tools and platforms rather than attempting a full genome-to-genome alignment in a basic script.")
    print("The pairwise alignment demonstrated above is a fundamental building block for such larger comparisons.")

# --- Example Usage ---
# IMPORTANT: Replace "your.email@example.com" with your actual email in the Entrez.email line above.

# 1. Basic analysis of a single gene (e.g., human TP53 mRNA)
target_accession_id_basic = "NM_000546.6"
perform_basic_sequence_analysis_from_ncbi(target_accession_id_basic)

# 2. Pairwise sequence alignment of two related human genes
# Example: Human TP53 mRNA (NM_000546.6) and Human TP63 mRNA (NM_003722.5)
# These are related genes in the p53 family.
seq1_for_alignment = "NM_000546.6" # Human TP53 mRNA
seq2_for_alignment = "NM_003722.5" # Human TP63 mRNA
perform_pairwise_sequence_alignment(seq1_for_alignment, seq2_for_alignment)

# 3. Conceptual explanation of genome comparison
conceptual_genome_comparison_explanation()
```

```python
# Load libraries
import GEOparse
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.stats.multitest as smm
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import gseapy as gp
from bioservices import UniProt
from goatools.obo_parser import GODag
from goatools.go_enrichment import GOEnrichmentStudy
from mygene import MyGeneInfo
```

```python
# Download GEO series
gse = GEOparse.get_GEO(geo="GSE10072", destdir=".", annotate_gpl=True)

# Extract expression data (genes √ó samples)
expr_df = gse.pivot_samples('VALUE')

# Load sample metadata
meta_df = gse.phenotype_data
display(expr_df.shape, meta_df.head())
```

```python
# Log2 transform if needed
expr_log = np.log2(expr_df + 1)

# Filter genes with low standard deviation
gene_sd = expr_log.std(axis=1)
expr_filt = expr_log[gene_sd > 0.05] # Lowering the threshold to 0.05
expr_filt.shape
```

```python
# Example: samples whose title contains 'control' vs 'treated'
control = meta_df[meta_df['title'].str.contains('Normal Lung')].index
treated = meta_df[meta_df['title'].str.contains('Lung Tumor')].index
```

```python
results = []
for gene in expr_filt.index:
    c_vals = expr_filt.loc[gene, control]
    t_vals = expr_filt.loc[gene, treated]
    t_stat, p_val = stats.ttest_ind(t_vals, c_vals, equal_var=False)
    results.append((gene, t_stat, p_val))

de_df = pd.DataFrame(results, columns=['gene','t_stat','p_value'])
de_df['adj_p'] = smm.multipletests(de_df['p_value'], method='fdr_bh')[1]
de_df['log2FC'] = expr_filt[treated].mean(axis=1) - expr_filt[control].mean(axis=1)
de_df = de_df.sort_values('adj_p')
display(de_df.head())
```

```python
# Select top DE genes
sig_genes = de_df[(de_df['adj_p'] < 0.05) & (abs(de_df['log2FC']) > 1)]['gene'].tolist()

# Enrichr enrichment
if sig_genes: # Add this check
    enr = gp.enrichr(gene_list=sig_genes,
                     gene_sets=['GO_Biological_Process_2021', 'KEGG_2021_Human'],
                     organism='Human',
                     outdir=None)
    go_bp = enr.results[enr.results['Gene_set'] == 'GO_Biological_Process_2021']
    kegg = enr.results[enr.results['Gene_set'] == 'KEGG_2021_Human']
    display(go_bp.head(), kegg.head())
else: # Add this else block
    print("Warning: No significant genes found for enrichment analysis.")
    go_bp = pd.DataFrame() # Define go_bp as an empty DataFrame
    kegg = pd.DataFrame() # Define kegg as an empty DataFrame
```

```python
# Map gene symbols to UniProt IDs
from bioservices import UniProt
from goatools.obo_parser import GODag
from goatools.go_enrichment import GOEnrichmentStudy

ugi = UniProt()
# Ensure sig_genes is not empty before mapping
if sig_genes:
    mapping = ugi.mapping(fr="GENENAME", to="ACC", query=",".join(sig_genes))
else:
    mapping = {}
    print("Warning: No significant genes found for UniProt mapping.")


godag = GODag("go-basic.obo", load_obsolete=False) # Load from local file

if mapping:
    # Extract UniProt IDs that have a mapping
    mapped_uniprot_ids = [v for k, v in mapping.items() if v] # Assuming mapping gives a dict of gene -> uniprot

    if mapped_uniprot_ids:
        goea = GOEnrichmentStudy(
            # Use all genes as population, assuming they are in expr_filt and thus mapped
            # A more rigorous approach would use a proper background population
            population=list(ugi.mapping(fr="GENENAME", to="ACC", query=",".join(expr_filt.index)) .values()),
            go2geneids=godag.get_go2geneids(mapped_uniprot_ids), # Need to map UniProt to GO IDs
            obodag=godag,
            propagate_counts=True,
            alpha=0.05,
            methods=['fdr_bh']
        )
        goea_results = goea.run_study(mapped_uniprot_ids)
        goea_df = pd.DataFrame([r.__dict__ for r in goea_results])
        display(goea_df[['name', 'p_fdr_bh', 'study_count']].head()) # Changed 'godag_term' to 'name'
    else:
        print("Warning: No UniProt IDs found for significant genes. Skipping GO enrichment.")
        goea_df = pd.DataFrame(columns=['name', 'p_fdr_bh', 'study_count']) # Create empty df to avoid NameError
else:
    print("Warning: No significant genes to map. Skipping GO enrichment.")
    goea_df = pd.DataFrame(columns=['name', 'p_fdr_bh', 'study_count']) # Create empty df to avoid NameError
```
